```python
#Method Overriding
class A:
    def add(self):
        print("This is a Classs A Method")
class B(A):
    def add(self):
        print("This is a Class B Method")
b=B()
b.add()
```

```python
#Operator Overloading
class A:
    def __init__(self,x,y):
        self.x=x
        self.y=y
    def __add__(self,value):
        return(self.x+value.x,self.y+value.y)
    def __lt__(self,value):
        return(self.x<value.y,self.x<value.y)
    def __gt__(self, value):
        return(self.x>self.x,value.y>value.y)
a=A(10,20)
b=A(20,30)
print(a+b)
print(a<b)
print(a>b)
```

```python
#Encapsulation :Binding members and Method into a Single Unit
#Class-->Data Members(Variables),Method(Functions)
#Abstraction:Data hiding
class A:
    __a=0
    def count(self):
        self.__a+=1
        print(self.__a)
obj=A()
obj.count()
obj.count()
print(obj._A__a)
```
