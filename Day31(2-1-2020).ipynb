# Turtle graphs

```python
# design spiral star
import turtle as t
a= t.Turtle()
a.pencolor('red')
for i in range(40):
    a.forward(i * 10)
    a.left(144)
t.done()    
```

```python
# design square
import turtle as t
a= t.Turtle()
for i in range(40):
    a.forward(i)
    a.left(90)
t.done()    
```

```python
# design spiral square
import turtle as t
a= t.Turtle()
for i in range(50):
    a.forward(i)
    a.left(92)
t.done()    
```
