# OOPs Concepts

```python
class Employee:
    name="Saiteja"
    rollno=543
    def display(self):
        print(self.name,self.rollno)
emp = Employee()
emp.display()
```

```python
# Using  Constructor:

class Student:
    def __init__(self,name,branch):
        self.name=name
        self.branch=branch
    def display(self):
        print(self.name,self.branch)
s = Student("Rajee","CSE")
s.display()
```

```python
#Counting the number of objects of a class

class A:
    c=0
    def __init__(self):
        A.c=A.c+1
a =A()
b =A()
print("number of objects: ",A.c)
```

```python
#Non-Parameterized Constructor:

class Frnd:
    def __init__(self):
        print("this is non parameterized construntor")
    def display(self,name):
        print("hello",name)
f=Frnd()
f.display("Akkkii")
```

```python
#Parameterized Constructor:

class Frnd:
    def __init__(self,name):
        print("this is parameterized")
        self.name=name
    def display(self):
        print("hello",self.name)
f=Frnd("Raja")
f.display()
    
```

```python
# Inheritance
# Single

class Animal:
    def speaking(self):
        print("Animal speaking")
class Dog(Animal):
    def barks(self):
        print("dog is barking")
b=Dog()
b.barks()
b.speaking()
```

```python
# Multiple 

class Animal:
    def Speaking(self):
        print("Animal is speaking")
class Dog(Animal):
    def barks(self):
        print("dog is barking")
class Cat(Animal):
    def meow(self):
        print("cat meow")
c=Cat()
b=Dog()
c.meow()
b.barks()
b.Speaking()
```

```python
# Multilevel

class Animal:
    def speaks(self):
        print("Animal is Speaking")
class Dog(Animal):
    def barks(self):
        print("Dog is Barking")
class Puppy(Dog):
    def eat(self):
        print("puppy is  eating")
d=Puppy()
d.barks()
d.speaks()
d.eat()
        
```
