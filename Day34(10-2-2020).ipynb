### Visualizing Data
  - An intresting way to display dataframe
  - With matplotlib

```python
import matplotlib.pyplot as plt
a1 = [1,2,3,4]
a2 = [5,6,7,8]
plt.figure(figsize=(4,4))
plt.plot(a1,a2,color='blue')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```
